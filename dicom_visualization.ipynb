```python
import pandas as pd
import numpy as np
import pydicom
import skimage
import matplotlib.pyplot as plt
```

```python
bbox = pd.read_csv('bounding_boxes.csv')
bbox
```

```python
dcm1 = pydicom.dcmread('dicom_00013659_019.dcm')
plt.imshow(dcm1.pixel_array, cmap='gray')
```

```python
plt.figure(figsize=(5,5))
plt.hist(dcm1.pixel_array.ravel(), bins=256)
plt.show
```

```python
x_mean = np.mean(dcm1.pixel_array)
x_std_dev = np.std(dcm1.pixel_array)
standard = (dcm1.pixel_array - x_mean)/x_std_dev
standard
```

```python
plt.hist(standard.ravel(), bins=256)
plt.show
```
